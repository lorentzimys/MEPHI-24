# Практическое задание №1
## Математическая модель оптимизации доставки для интернет-магазина

**Задача:** Интернет-магазин продает бытовую технику и электронику. У магазина есть склады в нескольких городах, так что он доставляет товары по всей стране. Магазин хочет минимизировать затраты на доставку и сократить время доставки до клиентов.
